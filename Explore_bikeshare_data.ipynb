### 1. Most Popular Hour of Day
> Determining the most popular hour of travel is crucial for Motivate. By analyzing the hour of travel data, we can identify the peak demand period and optimize resource allocation, bike availability, and rider experience. Insights gained from exploring temporal trends will help align operations, maintenance, and marketing strategies to cater to peak usage times, ensuring a seamless bike sharing experience that meets riders' preferences and travel habits.

```r
library(ggplot2)
library(dplyr)

# Create a temporary data frame with combined Start.Hour and End.Hour records
temp_df <- bind_rows(
  data.frame(Hour = combined_df$Start.Hour),
  data.frame(Hour = combined_df$End.Hour)
)

# Create a bar plot for month distribution
ggplot(temp_df, aes(x = Hour, fill = Hour)) +
  geom_bar() +
  labs(x = "Hour", y = "Count", title = "Most Popular Hour of Day") +
  scale_fill_manual(values = rainbow(length(unique(temp_df$Hour)))) +
  theme_minimal() +
  theme(legend.position = "none") +
  coord_flip()
```

#### Result Summary
>Based on the analysis of the dataset, it was found that the most popular hour of travel for Motivate is 8am, with a count of around 30,000 trips. This hour experiences a substantial surge in bike usage and represents the peak period of activity. In comparison, the next popular hour, 7am, has a count approximately 7,500 trips lower than the peak at 8am. The data reveals a clear pattern of high demand during the morning rush hour, indicating a strong preference for bike sharing as a convenient mode of transportation during this time.

>Following the 8am peak, there is a notable drop in trip counts until 5pm, when another peak occurs with a count of approximately 20,000 trips. While this count is 10,000 trips less than the 8am peak, it represents the next significant spike in usage after the morning hours. This finding suggests a shift in travel patterns throughout the day, with a dip in demand during the midday period and a resurgence in the late afternoon, likely driven by commuters returning from work or engaging in recreational activities. These insights provide valuable information for optimizing resource allocation and adjusting operational strategies to meet the demand fluctuations across different hours of the day.

### 3. Average Trip Duration (in minutes) by Gender

>Understanding the average trip duration by gender is essential potential address potential disparities or preferences that may exist. By examining the differences in trip durations between genders, Motivate can gain valuable insights into usage patterns and tailor their services to better meet the needs of all riders.

```r
# Filter the data for New York and Chicago cities and remove empty or NA values
filtered_df <- combined_df[combined_df$City %in% c("new york", "chicago"), ]

# Calculate the average trip duration in minutes for each gender in each city
avg_trip_duration <- aggregate(Trip.Duration/60 ~ City + Gender, data = filtered_df, FUN = mean)

# Filter the data to show only Female and Male genders
avg_trip_duration <- avg_trip_duration[avg_trip_duration$Gender %in% c("Female", "Male"), ]

# Create a grouped bar plot for average trip duration by city and gender
ggplot(avg_trip_duration, aes(x = City, y = `Trip.Duration/60`, fill = Gender)) +
  geom_bar(stat = "identity", position = "dodge") +
  labs(x = "City", y = "Average Trip Duration (minutes)", title = "Average Trip Duration by Gender") +
  theme_minimal()
```

#### Result Summary
> The analysis of average trip durations by gender in the cities serviced by Motivate reveals interesting findings. In Chicago, males had an average trip duration of approximately 13 minutes, while females had a slightly shorter duration of around 11 minutes. Similarly, in New York, males had an average trip duration of approximately 15 minutes, while females had a slightly shorter duration of around 12 minutes. These results highlight potential disparities or preferences in trip durations between genders. 

>By addressing these differences, Motivate can refine their services to better cater to the specific needs and preferences of male and female riders, promoting inclusivity and enhancing the overall customer experience.
